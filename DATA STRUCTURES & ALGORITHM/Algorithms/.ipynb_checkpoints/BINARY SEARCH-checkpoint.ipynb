numbers = [1,4,6,9,11,15,15,15,17,21,34,34,56]

number_to_find = 15

This should return 5,6,7 as indices containing number 15 in the array

```python
#binary search using single while loop
def binary_search(numbers_list, number_to_find):
    left_index = 0
    right_index = len(numbers_list) - 1
    mid_index = 0 
    
    while left_index <= right_index:
        mid_index = (left_index + right_index) // 2 
        mid_number = numbers_list[mid_index]
        
        if mid_number == number_to_find:
            return mid_index
        
        if mid_number < number_to_find: # this means number on the right side of the list
            left_index = mid_index + 1
            
        else: # number to find is on the left side of the list
            right_index = mid_index - 1
            
    return -1
    

if __name__ == '__main__':
    numbers_list = [1,4,6,9,10,5,7]
    sort_list = sorted(numbers_list)
    print(sort_list)
    number_to_find = 5
    
    index = binary_search(sort_list, number_to_find)
    print(f"Number found at index {index} using Binary search")
```

#####  This reason why it returns -1 index is because the numbers_list is an unsorted list so when you sort it, you would be able to find number 5. Binary Search required that list has to be sorted.

```python
#binary search using single while loop
def binary_search(numbers_list, number_to_find):
    left_index = 0
    right_index = len(numbers_list) - 1
    mid_index = 0 
    
    while left_index <= right_index:
        mid_index = (left_index + right_index) // 2 
        mid_number = numbers_list[mid_index]
        
        if mid_number == number_to_find:
            return mid_index
        
        if mid_number < number_to_find: # this means number on the right side of the list
            left_index = mid_index + 1
            
        else: # number to find is on the left side of the list
            right_index = mid_index - 1
            
    return -1

# Find index of all occurences of a nmber from sorted list
def find_all_occurences(numbers, number_to_find):
    index = binary_search(numbers, number_to_find)
    indices = [index]
    #find indices on left hand side
    i = index - 1
    while i >= 0:
        if numbers[i] == number_to_find:
            indices.append(i)
        else:
            break
        i -= 1
        
    # find indices on the righr hand side
    i = index + 1
    while i < len(numbers):
        if numbers[i] == number_to_find:
            indices.append(i)
        else:
            break
        i += 1
    
    return sorted(indices)


if __name__ == '__main__':
    numbers = [1,4,6,9,11,15,15,15,17,21,34,34,56]
    number_to_find = 15
    indices = find_all_occurences(numbers, number_to_find)
    print(f"Indices of occurances of {number_to_find} are {indices}")
```

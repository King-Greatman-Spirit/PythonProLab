A r-string is a dtring in python prefixed with a "r" and it tells python not to handle back slashes(' \ ') in any special way.
Normally backslashes are used to specify tabs, newline, e.t.c. For Example:

```python
print('\tTab')
```

```python
# Adding r-string

print(r'\tTab')
```
